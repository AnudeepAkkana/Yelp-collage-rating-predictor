```python
import pandas as pd

df_photos = pd.read_csv('Yelp/photos_df.csv').drop(columns='index')
df_business = pd.read_csv('Yelp/business_df.csv').drop(columns='index')
```

```python
df_photos.head()
```

```python
df_business.head()
```

```python
df_merged = df_business.merge(df_photos, left_on='business_id', right_on='business_id', how='inner', suffixes=['_b', '_p'])
# Removing the rows with label as 'menu'
df_merged = df_merged[df_merged['label'] != 'menu']
df_merged.head()
```

```python
import numpy as np

df_eda = df_merged.groupby('business_id')['stars', 'label', 'photo_id'].agg({
    'stars': np.mean, 
    'photo_id': 'count', 
    'label': 'nunique'
}).sort_values(by='label', ascending=False).rename(columns={
    'photo_id': 'number_of_photos',
    'stars': 'mean_stars'
}).reset_index()
df_eda
```

```python
import matplotlib.pyplot as plt

# Plotting the number photos
plt.plot(list(df_eda.index), df_eda['number_of_photos'])
plt.xlabel('Restaurant number')
plt.ylabel('Number of photos')
plt.show()

# Plotting number of labels
plt.plot(list(df_eda.index), df_eda['label'])
plt.xlabel('Restaurant number')
plt.ylabel('Number of labels')
plt.show()

# Plotting number of stars
plt.hist(df_eda['mean_stars'], bins=8)
plt.ylabel('Mean stars')
plt.show()
```

```python
business_ids = df_eda[df_eda['label'] == 4]['business_id']
print(len(business_ids))
df_photos_4_labels = df_merged[df_merged['business_id'].isin(list(business_ids))]
df_photos_4_labels.head()
```

```python
photo_index = df_photos_4_labels.groupby(['business_id', 'label', 'photo_id', 'stars'])['business_id'].agg('count')
photo_index
```

```python
# We are only using the images of businesses that have all 5 labels
# We are using a collage based CNN approach

# Step 1. Making the collages for testing and validation set
# Outer loop - businesses

import itertools
import cv2
import os
import random

photos = {}
for business_id in business_ids:
    photos['drink'] = list(photo_index[business_id, 'drink'].index)
    photos['food'] = list(photo_index[business_id, 'food'].index)
    photos['outside'] = list(photo_index[business_id, 'outside'].index)
    photos['interior'] = list(photo_index[business_id, 'inside'].index)
    values = list(photos.values())

    collage_pictures = list(itertools.product(*values))
    # Randomly shuffling the list for each business
    random.shuffle(collage_pictures)
    
    count = 0
    for picture in collage_pictures:

        # Limiting ourselves to 2 collages per business
        if count<=3:
            
            # Loading the pictures
            drink = cv2.imread(f'Yelp/photos/{picture[0][0]}.jpg', -1)
            food = cv2.imread(f'Yelp/photos/{picture[1][0]}.jpg', -1)
            outside = cv2.imread(f'Yelp/photos/{picture[2][0]}.jpg', -1)
            interior = cv2.imread(f'Yelp/photos/{picture[3][0]}.jpg', -1)

            average_sentiment = np.mean(np.array([picture[0][1], picture[1][1], picture[2][1], picture[3][1]]))
            
            # Resizing the pictures
            drink = cv2.resize(drink, (256, 256))
            food = cv2.resize(food, (256, 256))
            outside = cv2.resize(outside, (256, 256))
            interior = cv2.resize(interior, (256, 256))

            # Making the collages
            p1 = np.hstack((drink, food))
            p2 = np.hstack((outside, interior))
            collage = np.vstack((p1, p2))

            # Writing the files to the test class folders
            if os.path.exists(f"Yelp/collages/{average_sentiment}"):
                cv2.imwrite(f'Yelp/collages/{average_sentiment}/{business_id}_{collage_pictures.index(picture) + 1}.jpg', collage)
            else:
                os.makedirs(f'Yelp/collages/{average_sentiment}')
                cv2.imwrite(f'Yelp/collages/{average_sentiment}/{business_id}_{collage_pictures.index(picture) + 1}.jpg', collage)   
                
            count+=1           
```
